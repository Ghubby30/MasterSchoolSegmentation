```python
#############################      Data Frame with metrics      ##############
kset = kframe[['avg_hotel_occ_daily','total_bags','no_cancel_num','all_discounts','original_pay']]
```

```python
skaler = StandardScaler()
```

```python
###########Create K-Means************
kmeans = KMeans(n_clusters=5)
```

```python
kset[['avg_hotel_occ_daily_T','total_bags_T','no_cancel_num_T','all_discounts_T','original_pay_T']]=skaler.fit_transform(kset[['avg_hotel_occ_daily','total_bags','no_cancel_num','all_discounts','original_pay']])
kset
```

```python
df1 = kset.iloc[:,[6,7]].values
```

```python
# The other variables compared to "avg_hotel_occ_daily"
list1 = []
for k in range(1,6):
    kmeans =KMeans(n_clusters=k,init="k-means++")
    kmeans.fit(df1)
    list1.append(kmeans.inertia_)
plt.figure(figsize=(12,6))
plt.grid()
plt.plot(range(1,6), list1, linewidth=2, color="red",marker="*")
plt.xlabel("K-Value")
plt.ylabel("avg_hotel_occ_daily_T & total_bags_T")
plt.show()
```

```python
#Get labels
kmeans = KMeans(n_clusters=3)
label = kmeans.fit_predict(df1)
print(label)
```

```python
#Get centroids
print(kmeans.cluster_centers_)
```

```python
#Create scatterplot and plot the centroids
plt.scatter(df1[:,0], df1[:,1], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], color = 'black')
plt.title("Clusters of Customers")
plt.xlabel("avg_hotel_occ_daily")
plt.ylabel("total_bags")
plt.show()
```
